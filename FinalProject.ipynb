### Call the Socrata API to get Food Establishment results.
The code snippet below was taken from: https://dev.socrata.com/foundry/data.kingcounty.gov/gkhn-e8mn. Please visit for more information.

You can make limited calls without a token to explore. To operationalize though, it's recommended to get a token.

To get data for a different time range, simply change the dates in the where parameter.

```python
 
# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.kingcounty.gov", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.kingcounty.gov,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
#results = client.get("gkhn-e8mn", where="inspection_date > '2017-01-01'", limit=2000)
results = client.get("gkhn-e8mn", where="inspection_date > '2016-10-31' AND inspection_date < '2017-11-01'", limit=200000)
```

## Establishment Level Data Processing
### Get the counts for the 3 types of Inspection 
The Food Inspection dataset has mulitple rows for each establishment both for different types of inspections as well as violations. I need to extract the unique inspections. For our analysis, we only want to count inspections of type 'Routine Inspection/Field Review', so I needed to count each type of inspection separately. The output of the processing is available below the code cell. 

```python

# Filter to just a single Inspection type
InspectConsult = results_df[results_df['inspection_type'].isin(['Consultation/Education - Field'])]

# Select columns
InspectConsult = InspectConsult.iloc[:,[1,11]]

# Get counts after Group By
InspectConsult = InspectConsult.groupby("business_id").count()

# Re-add country as a column
InspectConsult["business_id"] = InspectConsult.index

# Set column names
InspectConsult.columns = ['InspectTypeCnt_Consult', 'business_id']

# Display sample
InspectConsult.head()



# Filter to just a single Inspection type
InspectRoutine = results_df[results_df['inspection_type'].isin(['Routine Inspection/Field Review'])]

# Select columns
InspectRoutine = InspectRoutine.iloc[:,[1,11]]

# Get counts after Group By
InspectRoutine = InspectRoutine.groupby("business_id").count()

# Re-add country as a column
InspectRoutine["business_id"] = InspectRoutine.index

# Set column names
InspectRoutine.columns = ['InspectTypeCnt_Routine', 'business_id']

# Display sample
InspectRoutine.head()



# Filter to just a single Inspection type
InspectReturn = results_df[results_df['inspection_type'].isin(['Return Inspection'])]

# Select columns
InspectReturn = InspectReturn.iloc[:,[1,11]]

# Get counts after Group By
InspectReturn = InspectReturn.groupby("business_id").count()

# Re-add country as a column
InspectReturn["business_id"] = InspectReturn.index

# Set column names
InspectReturn.columns = ['InspectTypeCnt_Return', 'business_id']

# Display sample
InspectReturn.head()



# Merge the population dataset with the total articles dataset.
myEstablishmentData = unique_bus.merge(InspectConsult, on='business_id', how='left')
myEstablishmentData["InspectTypeCnt_Consult"].fillna(0, inplace=True)

myEstablishmentData = myEstablishmentData.merge(InspectRoutine, on='business_id', how='left')
myEstablishmentData["InspectTypeCnt_Routine"].fillna(0, inplace=True)

myEstablishmentData = myEstablishmentData.merge(InspectReturn, on='business_id', how='left')
myEstablishmentData["InspectTypeCnt_Return"].fillna(0, inplace=True)

# Display a sample
myEstablishmentData.head()
```

### Get the counts for Red and Blue violations
Here I calculated the number of red and blue violations an establishment has. This is inclusive of both Routine and Return inspections.

```python
# Filter to just a single Violation type
ViolationBlue = results_df[results_df['violation_type'].isin(['blue'])]

# Select columns
ViolationBlue = ViolationBlue.iloc[:,[1,20]]

# Get counts after Group By
ViolationBlue = ViolationBlue.groupby("business_id").count()

# Re-add country as a column
ViolationBlue["business_id"] = ViolationBlue.index

# Set column names
ViolationBlue.columns = ['ViolationCnt_Blue', 'business_id']

# Display sample
ViolationBlue.head()



# Filter to just a single Violation type
ViolationRed = results_df[results_df['violation_type'].isin(['red'])]

# Select columns
ViolationRed = ViolationRed.iloc[:,[1,20]]

# Get counts after Group By
ViolationRed = ViolationRed.groupby("business_id").count()

# Re-add country as a column
ViolationRed["business_id"] = ViolationRed.index

# Set column names
ViolationRed.columns = ['ViolationCnt_Red', 'business_id']

# Display sample
ViolationRed.head()

# Merge Data
myEstablishmentData = myEstablishmentData.merge(ViolationBlue, on='business_id', how='left')
myEstablishmentData["ViolationCnt_Blue"].fillna(0, inplace=True)

myEstablishmentData = myEstablishmentData.merge(ViolationRed, on='business_id', how='left')
myEstablishmentData["ViolationCnt_Red"].fillna(0, inplace=True)

# Display a sample
myEstablishmentData.head()
```

### Calculate a Safety Group for each establishment
I wrote a function to calculate a SafetyGroup (simplified rating) for each establishment. This is similar to King County's simplified approach but is significantly different. Please see the code cell below for specific details.

```python
# Calculate a SafetyGroup for each establishment
def getInspectionGroup(row):
    if row['InspectTypeCnt_Routine'] == 0:
        val = 'NotInspected'
    elif row['ViolationCnt_Red'] >= 10:
        val = 'VeryLow'
    elif row['ViolationCnt_Red'] >= 5:
        val = 'Low'
    elif row['InspectTypeCnt_Routine'] > 2 and row['ViolationCnt_Red'] < 1:
        val = 'High'
    else:
        val = 'Average'
    return val
```

```python
# Call the funciton above to get the values
myEstablishmentData["SafetyGroup"] = myEstablishmentData.apply(getInspectionGroup, axis=1)
myEstablishmentData.head()
```

## City Level Data Processing

### Get Unique list of Cities and their population
I first get the list of unique cities from the Food Inspection dataset. I then merge with the population datset, keeping only keep records where the city is present in both datasets.

```python
# Get the list of unique cities from the Food Inspection dataset
unique_city = unique_bus.iloc[:,[2]].drop_duplicates()

# Merge with the population datset. We only keep records where the city is present in both datasets.
unique_city = unique_city.merge(pop_data, on='city', how='inner')

# Quick exploration of the output of the merge.
print("After merging, we have", unique_city.shape[0])
unique_city
```

### Calculate the number of establishments in each city
I calculate the number of establishments in each city and join the processed city level dataset.

```python
# Select columns
NumEstablisments = myEstablishmentData.iloc[:,[0,2]]

# Get counts after Group By
NumEstablisments = NumEstablisments.groupby("city").count()

# Re-add city as a column
NumEstablisments["city"] = NumEstablisments.index

# Set column names
NumEstablisments.columns = ['NumEstablisments', 'city']

# Merge datasets
unique_city = unique_city.merge(NumEstablisments, on='city', how='left')

# Display sample
unique_city.head()
```

### Calculate the number of Routine Inspections for each city
I calculate the number of routine inspections for each city and join the processed city level dataset.

```python
# Select columns
NumRoutineInspections = myEstablishmentData.iloc[:,[2,5]]

# Get counts after Group By
NumRoutineInspections = NumRoutineInspections.groupby("city").sum()

# Re-add city as a column
NumRoutineInspections["city"] = NumRoutineInspections.index

# Set column names
NumRoutineInspections.columns = ['NumRoutineInspections', 'city']

# Merge datasets
unique_city = unique_city.merge(NumRoutineInspections, on='city', how='left')

# Display sample
unique_city.head()
```

### Calculate the count and percent of High SafetyGroup Estabishments
For each city, I calculate the number of High SafetyGroup Estabishments and the percent of total establishments that number represents.

```python
# Filter to just the high safety group
SafetyGroupHigh = myEstablishmentData[myEstablishmentData['SafetyGroup'].isin(['High'])]

# Select columns
SafetyGroupHigh = SafetyGroupHigh.iloc[:,[0,2]]

# Get counts after Group By
SafetyGroupHigh = SafetyGroupHigh.groupby("city").count()

# Re-add city as a column
SafetyGroupHigh["city"] = SafetyGroupHigh.index

# Set column names
SafetyGroupHigh.columns = ['SafetyGroupCnt_High', 'city']

# Merge datasets
unique_city = unique_city.merge(SafetyGroupHigh, on='city', how='left')
unique_city["SafetyGroupCnt_High"].fillna(0, inplace=True)
unique_city["SafetyGroupPcnt_High"] = round(unique_city["SafetyGroupCnt_High"]*100/unique_city["NumEstablisments"],2)

# Display sample
unique_city.head()
```

### Calculate the count and percent of Lower SafetyGroup Estabishments
For each city, I calculate the number of Lower ("Low" and "Very Low") SafetyGroup Estabishments and the percent of total establishments that number represents.

```python
# Filter to just the low and very safety groups
SafetyGroupLower = myEstablishmentData[myEstablishmentData['SafetyGroup'].isin(['Low','VeryLow'])]

# Select columns
SafetyGroupLower = SafetyGroupLower.iloc[:,[0,2]]

# Get counts after Group By
SafetyGroupLower = SafetyGroupLower.groupby("city").count()

# Re-add city as a column
SafetyGroupLower["city"] = SafetyGroupLower.index

# Set column names
SafetyGroupLower.columns = ['SafetyGroupCnt_Lower', 'city']

# Merge datasets
unique_city = unique_city.merge(SafetyGroupLower, on='city', how='left')
unique_city["SafetyGroupCnt_Lower"].fillna(0, inplace=True)
unique_city["SafetyGroupPcnt_Lower"] = round(unique_city["SafetyGroupCnt_Lower"]*100/unique_city["NumEstablisments"],2)

# Display sample
unique_city.head()
```
